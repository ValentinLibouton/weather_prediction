```python
import matplotlib.pyplot as plt
import seaborn as sns
def display_nan_proportion(dataframe: pd.DataFrame, dataframe_name: str = "le dataframe"):
    nan_proportions = (dataframe.isna().sum() / len(dataframe)) * 100
    nan_proportions = nan_proportions.sort_values(ascending=False)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=nan_proportions.values, y=nan_proportions.index, palette='viridis')
    plt.xlabel('Proportion de NaN (%)')
    plt.title(f'Proportions de NaN par colonne dans {dataframe_name}')
    plt.gca().invert_yaxis()  # La colonne avec le plus de NaN en bas
    plt.show()

```

```python
display_nan_proportion(dataframe=weatherAUS, dataframe_name="weatherAUS")
```

```python
def fill_nan_with_mode(dataframe, column_name):
    """
    Remplace les valeurs NaN d'une colonne par la valeur la plus fréquente (mode).
    
    Args:
        dataframe (pd.DataFrame): Le DataFrame contenant la colonne à traiter.
        column_name (str): Le nom de la colonne à traiter.

    Returns:
        pd.DataFrame: Le DataFrame avec les valeurs NaN remplacées par le mode.
    """
    mode_value = dataframe[column_name].mode().values[0]
    dataframe[column_name].fillna(mode_value, inplace=True)
    return dataframe
```

```python
fill_nan_with_mode(weatherAUS, 'Sunshine')
fill_nan_with_mode(weatherAUS, 'Evaporation')
fill_nan_with_mode(weatherAUS, 'Cloud9am')
fill_nan_with_mode(weatherAUS, 'Cloud3pm')
fill_nan_with_mode(weatherAUS, 'Pressure9am')
fill_nan_with_mode(weatherAUS, 'Pressure3pm')
```

```python
def distribution_series(series:pd.Series):
    plt.figure(figsize=(10, 6))
    series.value_counts().sort_index().plot(kind='bar', color='skyblue')
    #plt.xlabel(series.name)
    plt.ylabel('Nombre de fois')
    plt.title(f'Distribution des valeurs de {series.name}')
    #plt.xticks(rotation=45)
    plt.show()
```
